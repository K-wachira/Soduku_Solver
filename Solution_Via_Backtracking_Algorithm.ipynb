# A function to print the board 

```python
def print_board(board):
    for i in range(len(board)): #row
        if i % 3 == 0 : print("______________________") # divided the rows with a line 
        for j in range(len(board[i])): #column
            if j % 3 == 0 and j != 0: print("|" , end = " ")
            print(board[i][j], end= ' ')
        print("") #breaks the columns accordingly 
```

# A function to get empty cells , detonated by 0

```python
def get_empty(board):
    for i in range(len(board)): #rows
        for j in range(len(board[i])): #columns
            if board[i][j] == 0:
                return i, j   #row , column  
    
    
    return None
```

# This function checks if the number can be inserted on the given position (pos) 

Checks if the given number is on the current row ,column or grid(3*3) square

```python
def check_if_valid(board, pos, num ):
    # pos is the position in the grid we want to enterthe number (row, column)
    # num is the number in question
    # check if num is already in the said row
  
    
    for i in range(len(board[0])): # checks the rows
        if board[pos[0]][i] == num and pos[1] != i :   # board[pos[0]] gets the row while [i] is the index of every number on the row 
            print(f"",num, " Cant be inserted as its on the rows")
            return False
        
    for i in range(len(board)):
        if board[i][pos[1]] == num and pos[0] != i:  # board[i] gets every row //column// while pos[1] gets the item at the column number given 
            print(f"",num, " Cant be inserted as its on the column")
            return False
        
    
    box_x = pos[0]//3  # gets the position of the grid vertically
    box_y = pos[1]//3 # gets the position of the grid horizontally
    
    
    for i in range(box_x* 3, box_x*3 +3):
        for j in range(box_y*3 , box_y*3 +3):
            if board[i][j] == num and (i,j) != pos:
                print(f"",num, " Cant be in as its in the grid ")
                return False
            
    return True
```

```python
def solver(board):
    is_empty = get_empty(board)
    
    if not is_empty: # meaninig the whole grid is filled 
        return True
    else:
        row, col = is_empty  # meaning there is an empty cell
    
    for i in range(1, 10): # loop though the numbers 1- 9
        
        if check_if_valid(board, (row, col), i): #check if the number is valid to be put in the cell 
            board[row][col] = i # if it is insert it 
            print(print_board(board))
            if solver(board): # call solver again with now the number inserted but check what it is returning , if true return true
                return True
            board[row][col]  = 0 #if it returns false reset the previous number
            
    return False
print(print_board(board))
print("\n************** Started ************\n")

solver(board) 

print("\n************** Done *****************\n")
print(print_board(board))
```
